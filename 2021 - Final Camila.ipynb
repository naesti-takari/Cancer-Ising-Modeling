##### Clase Cell <br>
Setea la información de cada posicion y contenido ("content"). Donde se considera que cuando "content" toma los siguientes valores: 
- "0" está vacía/empty

Además, setea la forma en que se imprime la clase.

```python
class Cell:
    def __init__(self, x, y, content):
        self.x = x
        self.y = y
        # El content debería ser: 0=empty, cancerCell=-1, immuneCell=1
        self.content = content

    def __str__(self):
        if self.content == 0:
            return "·"
        elif self.content == 1:
            return "o"
        elif self.content == -1:
            return "x"
        else:
            # raise InvalidStateException(self.content + " is not a recognized cell content")
            return "Hay un problema con la impresión de la Celula"

    def setContent(self, content):
        self.content = content
    # Puedo hacer una funcion dentro de la clase que  me diga qué tipo de vecinos tengo?
    # Si la posicion está en uun borde, va haber posiciones que no puede tomar, asi que imagino que tengo que devolver posiciones de los que podria tener vecinos para verlo.
```

##### Clase Board
- __createboard: 
A partir de darle la información del tamaño, genera un tablero de "Cell" vacías.
- getCellInPosition: Da la información de qué "Cell", está en la posición dada.
- getNeighbors: Da cuáles son las "Cell" que estan en las libertades de la posición dada.
- getNeighborsFromColor: Da cuáles son las "Cell" vecinas de cierto color.
- getBlankPositions: Da todas las posiciones vacías del tablero.

```python
class Board:
    def __init__(self, rows, cols):
        self.matrix = self.__createBoard(rows, cols)
        self.rows = rows
        self.cols = cols

    def __createBoard(self, rows, cols):
        matrix = []
        for i in range(rows):
            row = []
            for j in range(cols):
                row.append(Cell(i, j, 0))
            matrix.append(row)
        return matrix

    def getCellInPosition(self, x, y):
        return self.matrix[x][y]

    # Da los vecinos de la ficha que se quiere analizar.
    def getNeighbors(self, x, y):
        neighbors = []
        for i in range(x - 1, x + 2):
            for j in range(y - 1, y + 2):

                # Si la celda es la del medio, la ignoro
                if i == x and j == y:
                    # pdb.set_trace()
                    continue
                # Si la celda está en una esquina, la ignoro.
                elif i != x and j != y:
                    continue

                # Si se sale del tablero, la ignoro
                elif i < 0 or i >= self.rows or j < 0 or j >= self.cols:
                    continue

                neighbors.append(self.getCellInPosition(i, j))
        return neighbors

    # Me devuelve todos los vecinos de la posición, del color queq uiero analizar.
    def getNeighborsFromColor(self, x, y, colour):
        neighborsFromColour = []
        neighbors = self.getNeighbors(x, y)

        for d in neighbors:
            if d.content == colour:
                neighborsFromColour.append(d)
        return neighborsFromColour

    # Me da todas las posiciones vacías alrededor de la posición elegida
    def getBlankPositions(self):
        blankPositions = []

        for j in range(self.cols):
            for i in range(self.rows):
                currentPosition = self.getCellInPosition(i, j)
                if currentPosition.content == 0:
                    blankPositions.append(currentPosition)

        return blankPositions

    # Establezco cómo se va a imprimir el tablaro
    def __str__(self):
        board = ""
        for i in range(self.rows):
            row = ""
            for j in range(self.cols):
                row = row + str(self.getCellInPosition(i, j)) + " "
            board = board + row + '\n'
        return board
```

#### Definición de funciones
##### Función ijvalue
Da un par de valores i y j, para un k y nx definido.

```python
def ijvalue(k, nx):
    j = (np.floor(k/nx)).astype(int)
    i = (k - j*nx).astype(int)
    return i, j
```

##### Función getInteractionEnergies
Función que me da la energía de interacción para cada tipo de táctica, según la fuerza que tiene el sistema inmune y el sistema de células cancerosas.

```python
def getInteractionEnergies(cancerStrength="medium", immuneStrength="medium"):
    # Definición de la fuerza que tiene el sistema "canceroso"
    if cancerStrength == 'weak':
        cancerEye = 0.6
        cancerNet = 0.5
        cancerLadder = 0.4
        cancerConnection = 0.01
        cancerInvasion = 0.3

    elif cancerStrength == 'strong':
        cancerEye = 0.7
        cancerNet = 0.6
        cancerLadder = 0.5
        cancerConnection = 0.2
        cancerInvasion = 0.5

    # Si no es débil o fuerte, se toma que es medio.
    elif cancerStrength == "medium":
        cancerEye = 0.7
        cancerNet = 0.6
        cancerLadder = 0.5
        cancerConnection = 0.1
        cancerInvasion = 0.4
    else:
        print("Me estas pasando mal la fuerza del cancer, tiene que ser weak, medium o strong")

    # Definición de la fuerza que tiene el sistema inmune.
    if immuneStrength == 'weak':
        immuneEye = 0.3
        immuneNet = 0.6
        immuneLadder = 0.4
        immuneConnection = 0.01
        immuneInvasion = 0.4

    elif immuneStrength == 'strong':
        immuneEye = 0.5
        immuneNet = 0.8
        immuneLadder = 0.6
        immuneConnection = 0.2
        immuneInvasion = 0.6

    elif immuneStrength == "medium":    # Si no es débil o fuerte, se toma que es medio.
        immuneEye = 0.4
        immuneNet = 0.7
        immuneLadder = 0.5
        immuneConnection = 0.1
        immuneInvasion = 0.5
    else:
        print("Me estas pasando mal la fuerza del sistema inmune, tiene que ser weak, medium o strong")

#     energiesInteractionsCancer = [cancerEye, cancerNet, ancerLadder, cancerConnection, cancerInvasion]
#     energiesInteractionsImmune = [immuneEye, immuneNet, immuneLadder, immuneConnection, immuneInvasion]

    # Por una cuestión de simplicidad ahora, sólo voy a considerar la energía de interacción para las conecciones.
    energiesInteractionsCancer = [cancerInvasion, cancerConnection]
    energiesInteractionsImmune = [immuneInvasion, immuneConnection]

    return energiesInteractionsCancer, energiesInteractionsImmune
```

##### Función metropolis:
Evalúa si se hace el cambio según metropolis, tomando como argumento a la diferencia de energía y la temperatura de trabajo (en Kelvin). En el caso de poner la temperatura de trabajo en grados, podria tomar el valor "0" y no tiene solución.


```python
def metropolis(energyDiference, temperature=2):
    # Considero 310K de temperatura, que corresponde a 37ºC. T del cuerpo humano
    kBoltzmann = 1
#     kBoltzmann = 1.3806488E-23  # Unidades de J/K
    beta = 1 / (kBoltzmann * temperature)
    probability = np.exp(-energyDiference * beta)
    randomNumber = np.random.rand(1)

    if probability > randomNumber:
        return True
    elif probability < randomNumber:
        return False
```

##### Función getGroupSameColour:
Se fija todas las fichas que pertenecen al mismo grupo y color.

```python
def getGroupSameColour(board, cell):
    # Defino a groupArray como un set, es decir una "bolsita"
    groupArray = set([cell])
    queueToConsume = []
    queueToConsume.append(cell)

    while len(queueToConsume) > 0:
        current = queueToConsume.pop()
        currentNeighbors = board.getNeighborsFromColor(current.x, current.y, current.content)
        for neighbor in currentNeighbors:
            if neighbor not in groupArray:
                groupArray.add(neighbor)
                queueToConsume.append(neighbor)

    return groupArray
```

##### Función getEnergyDiference:
Función que calcula la diferencia de energía, dado el tablero, el jugador, la posicion que se quiere evaluar y las energias de interacción de los sistemas blanco y negro.

```python
def getEnergyDiference(board, player, currentCell, interactionEnergiesCancer, interactionEnergiesImmune):
    energyDiference = 0
    sumContent = 0
    zeroContent = 0
    groupSize = 0
    friendEnergy = 0
    enemyEnergy = 0
    groupsList = []
    cleanGroup = []

    # Solución para casos bordes
    colLimit = board.cols
    rowLimit = board.rows

    colsLimit = np.arange(colLimit)
    rowsLimit = np.arange(rowLimit)

    cornerCol = [0, colLimit - 1]
    cornerRow = [0, rowLimit - 1]

    xCur = currentCell.x
    yCur = currentCell.y

#     print("colLimit:", colLimit)
#     print("colsLimit", colsLimit)
# ------------------------------------------------------------------

    if player == "Blanco":
        cellType = 1
        friendInteraction = interactionEnergiesImmune
        enemyInteraction = interactionEnergiesCancer

    elif player == "Negro":
        cellType = -1
        friendInteraction = interactionEnergiesCancer
        enemyInteraction = interactionEnergiesImmune
    else:
        print("Me pasaste mal el jugador")

# ------------------------------------------------------------------

    # Analizo a los vecinos. Primero me fijo cuánto suman y si no tiene celulas vecinas.
    neighbors = board.getNeighbors(currentCell.x, currentCell.y)

    for d in neighbors:
        sumContent = sumContent + d.content
#         print("El vecino: fila=", d.x, " columna=", d.y, " contiene=", d.content)
        if d.content == 0:
            zeroContent = zeroContent + 1
        else:
            groupsList.append(getGroupSameColour(board, d))
#     print("Los vecinos vacíos son en total: ", zeroContent)
#     print("La suma del contenido es: ", sumContent)

    # Hago un unique, para poder quedarme con los grupos distintos y que no se repitan.
    cleanGroup = np.unique(groupsList)  # Momentaneo.

# ------------------------------------------------------------------

    # Me fijo si pertenece a la esquina.
    if xCur in cornerRow and yCur in cornerCol:
        liberty = 2

    # Me fijo si pertenece a un borde
    elif (xCur in rowsLimit and yCur in cornerCol) or (yCur in colsLimit and xCur in cornerRow):
        liberty = 3

    # Si no pertenece a nadie, puede tener 4 libertades
    else:
        liberty = 4

# ------------------------------------------------------------------

    # Considero que si solo hay vecinos vacíos, colocar la pieza es como una "invasion"
    if zeroContent == liberty:
        # currentCell.content = cellType  # Ojo, porque no tendria que asignar acá, debería ser en play.
        friendEnergy = friendInteraction[0]*cellType
#         print("Se coloca por zeroContent")

    else:
        for d in cleanGroup:
            # Trampita para acceder al contenido del set.
            for i in d:
                currentType = copy.copy(i).content
#                 print("Current Type", currentType)
#                 print("CellType", cellType)
            groupSize = len(d)
            if currentType == cellType:
                # Le sumo uno al groupSize, porque estoy poniendo la ficha y hace sinergia con elgrupo
                delta = friendInteraction[1]*cellType*(groupSize + 1) - friendInteraction[1]*cellType*groupSize
                friendEnergy = friendEnergy + delta
            else:
                enemyEnergy = enemyEnergy - enemyInteraction[1]*cellType*groupSize

    energyDiference = friendEnergy + enemyEnergy
#     test = board.getCellInPosition(currentCell.x, currentCell.y)
#     print("La celda queda: fila=", test.x, " columna=", test.y, " contiene=", test.content)
    return energyDiference, friendEnergy, enemyEnergy
```

##### Función play:


```python
def play(player, board, interactionEnergiesCancer, interactionEnergiesImmune):
    currFriendEnergy = 0
    currEnemyEnergy = 0
    currEnergyDiference = 0
    play = True
    playerPass = True
# ------------------------------------------------------------------
    if player == "Blanco":
        cellType = 1
    elif player == "Negro":
        cellType = -1
    else:
        print("Me pasaste mal el jugador")
# ------------------------------------------------------------------
    blankPositions = []
    # Se pide una posición random
    blankPositions = board.getBlankPositions()
    if len(blankPositions) == 0:
        play = False

    if play:
        currentCell = random.choice(blankPositions)
#         print("La posición que se está evaluando es: fila=", currentCell.x, " columna=", currentCell.y)
        energy = getEnergyDiference(board, player, currentCell, interactionEnergiesCancer, interactionEnergiesImmune)
        energyDiference = energy[0]
        friendEnergy = energy[1]
        enemyEnergy = energy[2]

        if cellType == 1:
            if energyDiference > 0:
                board.getCellInPosition(currentCell.x, currentCell.y).setContent(cellType)
                playerPass = False
            else:
                if metropolis(-energyDiference):
#                     print("Se hizo metropolis")
                    board.getCellInPosition(currentCell.x, currentCell.y).setContent(cellType)
                    playerPass = False
                else:
                    playerPass = True

        elif cellType == -1:
            if energyDiference < 0:
                board.getCellInPosition(currentCell.x, currentCell.y).setContent(cellType)
                playerPass = False
            else:
                if metropolis(energyDiference):
#                     print("Se hizo metropolis")
                    board.getCellInPosition(currentCell.x, currentCell.y).setContent(cellType)
                    playerPass = False
                else:
                    playerPass = True

        if not playerPass:
            currFriendEnergy = friendEnergy
            currEnemyEnergy = enemyEnergy
            currEnergyDiference = energyDiference

    return playerPass, currEnergyDiference, currFriendEnergy, currEnemyEnergy
```

```python
passes = 0
maxPasses = 2
totalPasses = 0
turn = 0
maxAnalyzeds = 10  # Setea la cantidad máxima de analisis de posiciones qeu va hacer cada jugador.
totalAnalyzeds = 0

acumulatedEnergyDiference = 0
acumulatedEnergyDiferenceList = []
variationEnergyDiference = []

acumulatedImmuneEnergy = 0
acumulatedImmuneEnergyList = []
variationImmuneEnergy = []

acumulatedCancerEnergy = 0
acumulatedCancerEnergyList = []
variationCancerEnergy = []

# ------------------------------------------------------------------

while passes < maxPasses:

    # Defino cuál es el player que juega. Si es par juega negro, si es impar juega blanco.
    if turn % 2 == 0:
        player = "Negro"

    elif turn % 2 != 0:
        player = "Blanco"
    else:
        print("Está mal puesto el tipo de player")
#    print("Juega", player)

    # Defino si hubo pases o no. Si no hubo pases, seteo a 0 la cantidad de pases (el juego termina cuando hay más de 2 pases seguidos)
    playerPass, energyDiference, friendEnergy, enemyEnergy = play(player, board, interactionEnergiesCancer, interactionEnergiesImmune)

#         print("Player:", player)
#         print("Energy Diference:", energyDiference)
#         print("FriendEnergy:", friendEnergy)
#         print("EnemyEnergy:", enemyEnergy)

    # Hago la acumulación de energía
    acumulatedEnergyDiference += energyDiference
    acumulatedEnergyDiferenceList.append(acumulatedEnergyDiference)
    variationEnergyDiference.append(energyDiference)

    if player == "Negro":
        # En realidad, no tengo variación de la energia enemiga, tengo solo de la amiga.
        acumulatedImmuneEnergy += 0
        acumulatedImmuneEnergyList.append(acumulatedImmuneEnergy)
        variationImmuneEnergy.append(0)

        acumulatedCancerEnergy += friendEnergy
        acumulatedCancerEnergyList.append(acumulatedCancerEnergy)
        variationCancerEnergy.append(friendEnergy)

    elif player == "Blanco":

        acumulatedImmuneEnergy += friendEnergy
        acumulatedImmuneEnergyList.append(acumulatedImmuneEnergy)
        variationImmuneEnergy.append(friendEnergy)

        acumulatedCancerEnergy += 0
        acumulatedCancerEnergyList.append(acumulatedCancerEnergy)
        variationCancerEnergy.append(0)

    else:
        print("Hay problemas con la asignación de Player")

    if playerPass:
        passes += 1
        totalPasses += 1
#         print("El jugador ", player, "pasó")
    elif not playerPass:
        passes = 0
#         print("El jugador", player, "jugó y el tablero quedó como\n")
#         print(board)
    else:
        print("Hay un problema con los pases")

    turn += 1
print("Hubo dos pases seguidos, terminó el juego.\nEl tablero final queda como:\n")
print(board)
```
